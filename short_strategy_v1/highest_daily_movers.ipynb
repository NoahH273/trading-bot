uses daily market data to create a sorted list with the tickers of the highest percent change stocks
input: 
    stocks: stock data in array format containg the percent change key
output:
    a sorted list of the highest moving stocks based on percentage change

```python
def filter_daily_data(stock_list, filters = []):
    def filter_check(stock):
        for x in filters:
            try:
                if stock[x] != None:
                    continue
            except KeyError:
                return False
        return True
        
    filtered_list = filter(filter_check, stock_list)
    filtered_list = list(filtered_list)
    return list(filtered_list)
```

```python
def get_highest_movers(stock_list, n = 50):
    def get_percent_change(stock):
        percent_change = stock['percent change']
        return percent_change
    stock_list.sort(reverse = True, key = get_percent_change)
    return stock_list[0:n]
```
